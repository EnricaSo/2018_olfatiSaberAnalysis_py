```python
# This script analyze the minimum distance among the agents in a 
# alpha alpha flock, as a function of the number of agents:
# d(N)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import csv

from mpl_toolkits.mplot3d import Axes3D
from numpy.linalg import norm


%matplotlib inline
```

```python
df = pd.read_csv('distance_data-N.csv')
```

```python
# Plot the distance for cc
# There is evidence of a decreasing trend

df_cut = df[df.ca==0.5][df.cc==0.5][df.cs==2][df.rep==1]
df_cut.reset_index(drop=True, inplace=True)
N = df_cut['N']

fig = plt.figure()
ax = plt.axes()
ax.set_xlabel('number of agents ($N$)')
ax.set_ylabel('distance [m]')
ax.set_title('Minimum distance among agents')

for k in N:
    ax.errorbar(k, np.squeeze(np.asarray(df_cut[df_cut.N==k]['avg_min_d'])), np.squeeze(np.asarray(df_cut[df_cut.N==k]['std_min_d'])), linestyle='None', marker='o', color='b');

plt.show()
```

```python
# Inverse polynomial system identification
import matplotlib.pyplot as plt
from matplotlib import pylab


# Scientific libraries
import numpy as np
from scipy.optimize import curve_fit

idx = np.argsort(N)
x = N[idx]
y = np.squeeze(np.asarray(df_cut['avg_min_d']))[idx]

def inverse_poly_func(x, a):
    return np.sqrt(a/x);

popt, pcov = curve_fit(inverse_poly_func, x, y)

x_new = np.linspace(0.5,200,200)
y_new = inverse_poly_func(x_new, *popt)

plt.plot(x, y, 'b', marker='o')
plt.plot(x_new, y_new, label='fit: a=%5.3f' % tuple(popt))
ax.set_xlabel('number of agents ($N$)')
ax.set_ylabel('distance [m]')
plt.title('Fit for minimum distance')
plt.legend()
plt.show()
```

```python
# Exponential system identification
import matplotlib.pyplot as plt
from matplotlib import pylab


# Scientific libraries
import numpy as np
from scipy.optimize import curve_fit

idx = np.argsort(N)
x = N[idx]
y = np.squeeze(np.asarray(df_cut['avg_min_d']))[idx]

def exponenial_func(x, a, b, c):
    return a*np.exp(-b*x)+c

popt, pcov = curve_fit(exponenial_func, x, y)

x_new = np.linspace(0.5,200,200)
y_new = exponenial_func(x_new, *popt)

plt.plot(x, y, 'b', marker='o')
plt.plot(x_new, y_new, label='fit: a=%5.3f b=%5.3f c=%5.3f' % tuple(popt))
ax.set_xlabel('number of agents ($N$)')
ax.set_ylabel('distance [m]')
plt.title('Fit for minimum distance')
plt.legend()
plt.show()
```

```python
# Plot maximum distance as a function of cc
fig = plt.figure()
ax = plt.axes()
ax.set_xlabel('number of agents ($N$)')
ax.set_ylabel('distance [m]')
ax.set_title('Maximum distance among agents')

for k in N:
    ax.plot(k, np.squeeze(np.asarray(df_cut[df_cut.N==k]['max_d'])), 'b', marker='o');

plt.show()
```
