```python
# This script analyze the minimum distance among the agents in a 
# Alpha Alpha flock, as a function of the number of agents:
# d(N)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import csv

from mpl_toolkits.mplot3d import Axes3D
from numpy.linalg import norm


%matplotlib inline
```

```python
df = pd.read_csv('distance_data-N.csv')
df = df.set_index('N')
df = df.sort_index()
```

```python
# Plot the inter-agent distance with std-dev
# for N <= 10

N = df.index.values

cut_at_idx = 11
idx_list = df.index.values < cut_at_idx
idx_values = df[idx_list].index.values

fig = plt.figure()
ax = plt.axes()

for k in idx_values:
    ax.errorbar(k, np.squeeze(np.asarray(df.ix[k,'avg_min_d'])), np.squeeze(np.asarray(df.ix[k,'std_min_d'])), 
                linestyle='None', marker='o', color='b');

ax.set_xlabel('Number of agents $N$ [#]')
ax.set_ylabel('Inter-agent distance [m]')

plt.ylim([1.3,2.1])

plt.savefig('figures/dist_std_zoom.png', bbox_inches='tight')
plt.show()
```

```python
# Plot the inter-agent distance with min-max
# for N <= 10

fig = plt.figure()
ax = plt.axes()

plot1, = plt.plot(idx_values, df.ix[idx_list,'min_min_d'],'ro')
plot2, = plt.plot(idx_values, df.ix[idx_list,'max_min_d'],'go')
plot3, = plt.plot(idx_values, df.ix[idx_list,'avg_min_d'],'bo')

ax.set_xlabel('Number of agents $N$ [#]')
ax.set_ylabel('Inter-agent distance [m]')
ax.legend((plot2, plot3, plot1), ('max', 'average', 'min'))

plt.ylim([1.3,2.1])

plt.savefig('figures/dist_min_max_zoom.png', bbox_inches='tight')
plt.show()
```

```python
# Plot the inter-agent distance with std-dev
# for N > 10

cut_at_idx = 10
idx_list = df.index.values > cut_at_idx
idx_values = df[idx_list].index.values

fig = plt.figure()
ax = plt.axes()

for k in idx_values:
    ax.errorbar(k, np.squeeze(np.asarray(df.ix[k,'avg_min_d'])), np.squeeze(np.asarray(df.ix[k,'std_min_d'])), 
                linestyle='None', marker='o', color='b');

ax.set_xlabel('Number of agents $N$ [#]')
ax.set_ylabel('Inter-agent distance [m]')

plt.ylim([1.3,2.1])

plt.savefig('figures/dist_std.png', bbox_inches='tight')
plt.show()
```

```python
# Plot the inter-agent distance with min-max
# for N > 10

fig = plt.figure()
ax = plt.axes()

plot1, = plt.plot(idx_values, df.ix[idx_list,'min_min_d'],'ro')
plot2, = plt.plot(idx_values, df.ix[idx_list,'max_min_d'],'go')
plot3, = plt.plot(idx_values, df.ix[idx_list,'avg_min_d'],'bo')

ax.set_xlabel('Number of agents $N$ [#]')
ax.set_ylabel('Inter-agent distance [m]')
ax.legend((plot2, plot3, plot1), ('max', 'average', 'min'))

plt.ylim([1.3,2.1])

plt.savefig('figures/dist_min_max.png', bbox_inches='tight')
plt.show()
```
