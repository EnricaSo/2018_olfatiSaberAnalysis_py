```python
# This script analyze the minimum distance among the agents in a 
# alpha alpha flock, as a function of the number of agents:
# d(N)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import csv

from mpl_toolkits.mplot3d import Axes3D
from numpy.linalg import norm


%matplotlib inline
```

```python
df = pd.read_csv('distance_data_N.csv')
```

```python
# Plot the distance
# There is evidence of a constant trend

N = df['N']

fig = plt.figure()
ax = plt.axes()
ax.set_xlabel('number of agents ($N$)')
ax.set_ylabel('distance [m]')
ax.set_title('Minimum distance among agents')

for k in N:
    ax.errorbar(k, np.squeeze(np.asarray(df[df.N==k]['avg_min_d'])), np.squeeze(np.asarray(df[df.N==k]['std_min_d'])), linestyle='None', marker='o', color='b');

plt.show()
```
